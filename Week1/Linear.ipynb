```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get data from the dataset
data_file = "LifeExpectancyDataset - Sheet1.csv"
df_og = pd.read_csv(data_file)
```

```python
df_og = df_og.dropna()
df = df_og.sample(frac=1, random_state=42).reset_index(drop=True)
# plt.figure(figsize=(8, 6))
# plt.scatter(df["Alcohol"], df["Life expectancy"], color='blue', marker='o')
```

```python
y_col = "Life expectancy"
string_cols = ["Country", "Year", "Status"]
Y = df[y_col].to_numpy()
X = df.drop(columns=string_cols + [y_col]).to_numpy()
X = np.c_[X, np.ones(X.shape[0])]
```

```python
split = int(0.8 * X.shape[0])
X_train, X_test = X[:split], X[split:]
Y_train, Y_test = Y[:split], Y[split:]
```

```python
def rmse(w, X, Y):
    m = X.shape[0]
    cost = 0
    for i in range(m):
        cost += (X[i] @ w - Y[i]) ** 2
    cost = cost / (2*m)
    return np.sqrt(cost)
```

```python
test_loss = rmse(w, X_test, Y_test)
print(test_loss)
```
